### 
Converting this to different format 
-  [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'----------------["Action","Adeventure","Fantasy","Fiction"]

```python
# as the whole line of genres are string we have make it in list of string for this we are using ast.literal_eval which wil make it int olist of string
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
#example
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies["genres"]=movies["genres"].apply(convert)
#now through this we succesfully converted the genres into our liking
```

```python
movies.head()
```

```python
#now performing the same thing to keywords column
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
movies.iloc[1]["cast"]
```

```python
def top3(obj):
    j=[]
    for i in ast.literal_eval(obj):
        j.append(i["name"])
        if(len(j)==3):
            return j
            break
    return j
        
```

```python
movies["cast"]=movies["cast"].apply(top3)
```

```python
movies.head()
```

```python
movies["crew"][1]
```

```python
import ast
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

```python
#as we know that the context in overview are in the form of string
movies["overview"][0]
#so for convertring it to list of string so that we can later concatinate it with other columns we are gonna use lambda function
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

-  Now the problem is that there could multiple similar first names but there second name would be different so afte the token is been generated the first name would be gnereated differnetly and second name would separted so for that we have to removes  the white spaces from the name so that the whole name will be one entity

## Text to vector

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=500,stop_words="english")
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()
print(f"The shape of the matrices is {vectors.shape}\n{vectors[0]}")
```

```python
cv.get_feature_names()
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
def Stem(obj):
    k=[]
    for i in obj.split():
        k.append(ps.stem(i))
    return " ".join(k)
```

```python
new_df["tags"]= new_df["tags"].apply(Stem)
```

```python
new_df["tags"][0]
```

```python
cv.get_feature_names()
```

Then after this we for finding the similarities between the distances we could either find the eucledian distance method to find the clooseness but for this case we are considering to find <b>COSINE DISTANCE </b>
-  Cosine distance basically finds the angle between two points whoever has the clossest angle will likely to have similarites
-  Another reason for considering COSINE DISTANCE is that the higher the dimensional data heigher the chances of fails for getting similarites using Eucledian method 

```python
from sklearn.metrics.pairwise import cosine_similarity
similar=cosine_similarity(vectors)
```

Now as we have find the values whcih are similar to the target movie for that we have to sort the matrices so that it will give the top similar movie but becuase of that we will loose the index of the movie which has the similarites of the targeted  movie 
for that we will using <b>ENUMERATE</b> function whcih will assign an index starting from zero

The purpose of this code is to sort a list of similarities (or scores) in descending order, while preserving the original indices. This is commonly used in tasks like finding the most similar items, such as in a recommendation system.
-  <b>Enumarate</b>-This function adds a counter to the similar[0] list and returns it as an enumerate object.
-  <b>Enumarate</b>-This step ensures that you have a list of tuples that can be easily manipulated or sorted.
-  <b>sorted(..., key=lambda x: x[1])</b>-lambda x: x[1]: A lambda function is an anonymous function defined with the keyword lambda. In this case, it takes a tuple x and returns its second element x[1]
-  <b>reverse</b>- It is used to sort the array in descending order

```python
storing=sorted(list(enumerate(similar[0])),reverse=True,key=lambda x:x[1])[1:6]
storing [2][0]
```
